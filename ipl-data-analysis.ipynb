```python
def make_team(team_squad):
    team=team_squad.copy()
    no_of_batsmen=4
    no_of_batting_allrounders=3
    no_of_bowling_allrounders=0
    no_of_bowlers=4
    foreigners=4
    batting_all_list=[]
    bat_list=[]
    bowl_list=[]
    bowling_all_list=[]
    playing11=[]
    fore=[]
#     batting_all=team_squad[team_squad.Cluster==3]
#     bowling_all=team_squad[team_squad.Cluster==2]
#     batter=team_squad[team_squad.Cluster==1]
#     bowler=team_squad[team_squad.Cluster==0]
    wk=1
    foreigners_in=team_squad[team_squad.Country_Name=='Overseas']
    foreigners_in=foreigners_in.sort_values(by=['Distance'])
    
#     picked=foreigners.iloc[:4,:]
    while foreigners>0:
#         print(foreigners_in)
        player=foreigners_in.iloc[0,0]
#         print(player)
        cluster=foreigners_in.iloc[0,24]
        role=foreigners_in.iloc[0,36]
        if role=='Wicket-Keeper':
            wk-=1
        if cluster==0:
            if no_of_bowlers>0:
                no_of_bowlers-=1
                bowl_list.append(player)
                foreigners-=1
                fore.append(player)
        elif cluster==1:
            if no_of_batsmen>0:
                no_of_batsmen-=1
                bat_list.append(player)
                foreigners-=1
                fore.append(player)
        elif cluster==2:
            if no_of_bowling_allrounders>0:
                no_of_bowling_allrounders-=1
                bowling_all_list.append(player)
                foreigners-=1
                fore.append(player)
        else:
            if no_of_batting_allrounders>0:
                no_of_batting_allrounders-=1
                batting_all_list.append(player)
                foreigners-=1
                fore.append(player)
#         playing11.append(row['player_name'])
        player_index=foreigners_in.index[foreigners_in['player_name']==player].tolist()
#       print(player)
        foreigners_in=foreigners_in.drop(player_index)
#         picked.append(player)
#         foreigners_in=foreigners_in.iloc[1:,:]
        if foreigners_in.empty:
            break
        
    #Only Indian Players
    indians=team_squad[team_squad.Country_Name=='India']
    #batting all rounders
    batting_all=indians[indians.Cluster==3]
    batting_all=batting_all.sort_values(by=['Distance_c3'])
    for i in range(no_of_batting_allrounders):
        player=batting_all.iloc[0,0]
        player_index=batting_all.index[batting_all['player_name']==player].tolist()
#       print(player)
        indians=indians.drop(player_index)
        batting_all_list.append(player)
        batting_all=batting_all.iloc[1:,:]
        #print(batting_all)
        if batting_all.empty:
            break
    #batsmen
    batsman=indians.copy()
    batsman=batsman.sort_values(by=['Distance_c1'])
    for i in range(no_of_batsmen-1):
        player=batsman.iloc[0,0]
        role=batsman.iloc[0,36]
        if wk>0:
            if role=='Wicket-Keeper':
                wk-=1
        player_index=batsman.index[batting_all['player_name']==player].tolist()
#       print(player)
        indians=indians.drop(player_index)
        bat_list.append(player)
        batsman=batsman.iloc[1:,:]
        #print(batting_all)
        if batsman.empty:
            break
    if wk>0:
        batsman=batsman[batsman.Bowling_skill=='Wicket-Keeper']
        player=batsman.iloc[0,0]
        player_index=batsman.index[batting_all['player_name']==player].tolist()
#       print(player)
        indians=indians.drop(player_index)
        bat_list.append(player)
        batsman=batsman.iloc[1:,:]
    else:
        player=batsman.iloc[0,0]
        player_index=batsman.index[batting_all['player_name']==player].tolist()
#       print(player)
        indians=indians.drop(player_index)
        bat_list.append(player)
        batsman=batsman.iloc[1:,:]
    no_of_batsmen-=1
        
    #bowling all rounders
    bowling_all=indians[indians.Cluster==2]
    bowling_all=bowling_all.sort_values(by=['Distance_c2'])
    for i in range(no_of_bowling_allrounders):
        player=bowling_all.iloc[0,0]
        player_index=bowling_all.index[batting_all['player_name']==player].tolist()
#       print(player)
        indians=indians.drop(player_index)
        bowling_all_list.append(player)
        batting_all=batting_all.iloc[1:,:]
        #print(batting_all)
        if bowling_all.empty:
            break
    #bowlers
    bowlers=indians.copy()
    bowlers=bowlers.sort_values(by=['Distance_c0'])
    for i in range(no_of_bowlers):
        player=bowlers.iloc[0,0]
        player_index=bowlers.index[batting_all['player_name']==player].tolist()
#       print(player)
        indians=indians.drop(player_index)
        bowl_list.append(player)
        bowlers=bowlers.iloc[1:,:]
        #print(batting_all)
        if bowlers.empty:
            break
    selected=[]
    selected.append(batting_all_list)
    selected.append(bowling_all_list)
    selected.append(bowl_list)
    selected.append(bat_list)
    print('batting_all',batting_all_list,'bowling_all',bowling_all_list,'bowl',bowl_list,'bat',bat_list)
    
            
    
    
    
    
    
        
```
